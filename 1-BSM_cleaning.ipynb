```python
import json

byte_data_list = []
for index, row in BSM1_dataset_scaffolded.iterrows():
    # Convert the row to a dictionary
    row_dict = row.to_dict()
    # Convert the dictionary to JSON string
    row_json = json.dumps(row_dict)
    # Encode the JSON string into bytes using UTF-8 encoding
    row_bytes = row_json.encode('utf-8')
    # Append the bytes representation to the list
    byte_data_list.append(row_bytes)

# Print the list of byte data
for i, byte_data in enumerate(byte_data_list):
    print(f"Row {i + 1} as bytes:", byte_data)
```

```python
import struct
import csv
from cryptography.hazmat.primitives.padding import PKCS7
from cryptography.hazmat.primitives.ciphers import Cipher, algorithms, modes
import random
import string

# Define the format string for packing the BSM data into bytes
bsm_format = "3s 19s i i i i i i i i i i i i i i i i i i i"

# Convert each row of the DataFrame into bytes
byte_data_list = []
for index, row in BSM1_dataset_scaffolded.iterrows():
    # Pack the BSM data into bytes using struct.pack
    bsm_bytes = struct.pack(bsm_format,
                             row['dataType'].encode('utf-8'),
                             row['metadata_generatedAt'].encode('utf-8'),  
                             row['coreData_position_lat'],
                             row['coreData_position_long'],
                             row['coreData_elevation'],
                             row['coreData_accelset_accelYaw'],
                             row['coreData_accelset_lat'],
                             row['coreData_accelset_long'],
                             row['coreData_accelset_vert'],
                             row['coreData_angle'],
                             row['coreData_accuracy_orientation'],
                             row['coreData_accuracy_semiMajor'],
                             row['coreData_accuracy_semiMinor'],
                             row['coreData_speed'],
                             row['coreData_heading'],
                             row['coreData_brakes_wheelBrakes_leftFront'],
                             row['coreData_brakes_wheelBrakes_rightFront'],
                             row['coreData_brakes_wheelBrakes_leftRear'],
                             row['coreData_brakes_wheelBrakes_rightRear'],
                             row['Width'],
                             row['Length'])
    byte_data_list.append(bsm_bytes)

filename = 'bsm.csv'
rows = []

# Print the list of byte data
for i, byte_data in enumerate(byte_data_list):
    # Convert each byte to its hexadecimal representation
    hex_data = [hex(b) for b in byte_data]
    # Remove the '0x' prefix from each hex value
    formatted_data = [h[2:] for h in hex_data]
    # Join the formatted data with commas and quotes
    output_string = ''.join(formatted_data)
    N = len(output_string)
    if N%16 != 0:
        pad_length = 16 - (N%16)
        for x in range(pad_length):
            output_string += random.SystemRandom().choice(string.ascii_lowercase + string.digits)
    extra_length = i*16
    for y in range(extra_length):
        output_string += random.SystemRandom().choice(string.ascii_lowercase + string.digits)
    output_string = output_string[:-1]
    row = []
    print(output_string)
    print(len(output_string))
    print("------------------------")
    row.append(output_string)
    rows.append(row)




with open(filename, 'w') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(['BSM hex bytes'])
    csvwriter.writerows(rows)
```
